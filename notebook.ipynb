```python
News_dic=dict()

News_dic['train']={
"Ejtemaee":[],
"Adian":[],
"Eghtesasdi":[],
"Siasi":[],
"Fanawari":[],
"Rahbord":[],
"Warzesh":[]}

News_dic['test']={
"Ejtemaee":[],
"Adian":[],
"Eghtesasdi":[],
"Siasi":[],
"Fanawari":[],
"Rahbord":[],
"Warzesh":[]}

```

```python
def open_file(*args):
    for dirpath, dirnames, filenames in os.walk(*args):
        for filename in filenames:
            filepath = os.path.join(dirpath, filename)
            yield (filepath, open(filepath, "r", encoding='utf-8'))

```

```python
mydir = "./Final_Dataset/Train"

for filepath, file in open_file(mydir):
    if 'اجتماعی' in filepath:
        News_dic['train']["Ejtemaee"].append(file.read())
        
    elif 'اديان' in filepath:
        News_dic['train']["Adian"].append(file.read())

    elif 'اقتصادی' in filepath:
        News_dic['train']["Eghtesasdi"].append(file.read())

    elif 'سیاسی' in filepath:
        News_dic['train']["Siasi"].append(file.read())

    elif 'فناوري' in filepath:
        News_dic['train']["Fanawari"].append(file.read())

    elif 'مسائل راهبردي ايران' in filepath:
        News_dic['train']["Rahbord"].append(file.read())

    elif 'ورزشی' in filepath:
        News_dic['train']["Warzesh"].append(file.read())

    else:
        continue;
        
```

```python
Most_words_each_label=dict()
Most_words_each_label={
  "Ejtemaee":{},
"Adian":{},
"Eghtesasdi":{},
"Siasi":{},
"Fanawari":{},
"Rahbord":{},
"Warzesh":{}}  



for keys in Most_words_each_label.keys():
    for news in News_dic['train'][keys]:
        news_splited=word_tokenize(news)
        for word in news_splited:
            if word not in stop_words:
                word=word.translate(str.maketrans('', '', string.punctuation))

                if '\u200c' in word:
                    word=word.split('\u200c')[0]
                if '\u200f' in word:
                    word=word.split('\u200c')[0] 

                if 'amp' in word:
                    continue
                if 'nbsp' in word:
                    continue
                if word in Most_words_each_label[keys].keys():
                    Most_words_each_label[keys][word]+=1
                else:
                    Most_words_each_label[keys][word]=1



#     Most_words_each_label[keys][kalame]


print((Most_words_each_label['Ejtemaee']))

```

```python
for key in Most_words_each_label.keys():
    Most_words_each_label[key]=dict(sorted(Most_words_each_label[key].items(), key=lambda item: item[1]))
    print(key , 'length = ' , len(Most_words_each_label[key]))
    
print((Most_words_each_label['Ejtemaee']))
```

```python
five_hundred_list=[]
for key in five_hundred.keys():
    five_hundred_list.append(key)
five_hundred_list=set(five_hundred_list)
print('برق' in five_hundred_list)

```

```python
array_news_500word_train=dict()
number=-1


for keys in Most_words_each_label.keys():
    for news in News_dic['train'][keys]:
        number+=1
        array_news_500word_train[number]=[0] * 500
        news_splited=word_tokenize(news)
        for word in news_splited:
            word=word.translate(str.maketrans('', '', string.punctuation))
            if '\u200c' in word:
                word=word.split('\u200c')[0]
            if 'amp' in word:
                continue
            if 'nbsp' in word:
                continue
            if word in five_hundred_list:
                word_id=five_hundred[word]
                array_news_500word_train[number][word_id]+=1

print(array_news_500word_train)
```

```python
array_news_500word_test=dict()
number=-1


for keys in Most_words_each_label.keys():
    for news in News_dic['test'][keys]:
        news_splited=word_tokenize(news)
        number+=1
        array_news_500word_test[number]=[0] * 500
        for word in news_splited:
            word=word.translate(str.maketrans('', '', string.punctuation))
            if '\u200c' in word:
                word=word.split('\u200c')[0]
            if 'amp' in word:
                continue
            if 'nbsp' in word:
                continue
            if word in five_hundred_list:
                word_id=five_hundred[word]
                array_news_500word_test[number][word_id]+=1
```
